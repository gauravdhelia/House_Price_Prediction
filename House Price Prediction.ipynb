Inserting the null values with 0 for columns containing numeric data type

```python
print(train_df.select_dtypes(include=np.number).columns)
for i in ('MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual', 'OverallCond',
       'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
       'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
       'MoSold', 'YrSold'):
    train_df[i]=train_df[i].fillna(0)
    test_df[i]=test_df[i].fillna(0)

```

Inserting the null values with 'None' for columns containing non-numeric data type

```python
print(train_df.select_dtypes(exclude=np.number).columns)
for i in ('MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
       'PavedDrive', 'SaleType', 'SaleCondition'):
    train_df[i]=train_df[i].fillna('None')
    test_df[i]=test_df[i].fillna('None')
                                
```

Creating a new column 'train' in the dataframe in which '1' is assigned to the training data and '0' to the test data. Subsequently applying One-Hot encoding as our data contains a number of categorical data types. The new column has been created to apply One-Hot encoding simultaneously to both the train and test data. The new column has been created to easily separate the train and test data after the application of One-Hot encoding.

```python
train_df['train']=1
test_df['train']=0
combine=pd.concat([train_df,test_df])
```

```python
ohe_df=pd.get_dummies(combine,columns=['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
       'PavedDrive', 'SaleType', 'SaleCondition'])
```

```python
#Segregating the data after One-Hot encoding and dropping the 'SalePrice' column created in the test data (null filled)
train_df=ohe_df[ohe_df['train']==1]
test_df=ohe_df[ohe_df['train']==0]
train_df.drop(['train'],axis=1,inplace=True)
test_df.drop(['train','SalePrice'],axis=1,inplace=True)
```

```python
#Creating inputs for model application
X_train=train_df.drop('SalePrice',axis=1)
Y_train=train_df['SalePrice']
X_test=test_df
```

```python
#Applying Linear Regression to our Regression Problem
linreg=LinearRegression()
linreg.fit(X_train,Y_train)
Y_pred=linreg.predict(X_test)
```

```python
#Creating a submission dataframe and finally saving our submission .csv file 
submission=pd.DataFrame({"Id": test_parent.Id, "SalePrice": Y_pred})
print(submission)
submission.to_csv('house_submission.csv')
```
